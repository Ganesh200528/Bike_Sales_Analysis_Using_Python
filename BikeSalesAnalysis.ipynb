# **This Project About : Bike Sales Analysis**


#### **🎯 Objective of Bike Sales Analysis**
The objective of the Bike Sales Analysis is to examine sales data to gain meaningful insights into customer behavior, product performance, and financial outcomes. By analyzing key variables such as date, customer demographics, location, product categories, and sales metrics, the goal is to:

Identify sales trends over time (daily, monthly, yearly)

Understand customer preferences based on age, gender, and region

Determine the best-selling products and most profitable categories

Evaluate the effectiveness of pricing strategies (unit cost vs. unit price)

Analyze revenue, cost, and profit patterns to assess overall business health

Support data-driven decisions for marketing, inventory, and sales strategies

Ultimately, the analysis aims to improve customer targeting, optimize inventory, boost profitability, and guide strategic planning for business growth.

#### **🏢 About the Company**
The company is a bicycle retailer and distributor that offers a wide range of bikes and cycling accessories, including road bikes, mountain bikes, hybrids, and electric models. It operates across multiple countries and states, serving a diverse customer base of various age groups and preferences.
With a focus on quality, customer satisfaction, and sustainable transportation, the company uses data-driven strategies to monitor sales performance, understand customer behavior, and improve decision-making in areas such as marketing, inventory management, and product development.
#### **Dataset Overview** 
The dataset includes the following columns:  
- **`date`**: Full date of purchase
- **`day`**: Day of the week (e.g., Monday, Tuesday)
- **`month`**: Month of the year
- **`year`**: Year of the transaction
- **`customer_age`**: Age of the customer
- **`age_group`**: Age category 
- **`customer_gender`**: Gender of the customer
- **`country`**: Country where the purchase was made
- **`state`**: State or region of the sale
- **`product_category`**: Main category of the bike or accessory
- **`sub_category`**: Specific type within the product category
- **`products`**: Name or description of the sold product
- **`order_quantity`**: Number of units ordered
- **`unit_cost`**: Cost to the company per unit
- **`unit_price`**: Selling price per unit
- **`profit`**: Net profit from the sale
- **`cost`**: Total cost of the order
- **`revenue`**: Total income from the sale

```python
# Calculate total and average Profit
total_profit = df.Profit.sum()
print("Total Profit:", total_profit)

average_profit = df.Profit.mean()
print("Average Profit:", average_profit)

# Calculate total and average Revenue
total_revenue = df.Revenue.sum()
print("Total Revenue:", total_revenue)

average_revenue = df.Revenue.mean()
print("Average Revenue:", average_revenue)

```

```python
# in that may give the unique values of country
df['Country'].unique()
```

```python
#profit by country wisely
sns.scatterplot(x="Country",y="Profit",data= df)
plt.show()
```

```python
# Scatter plot of Revenue vs Profit, colored by Customer Gender
sns.scatterplot(x="Revenue", y="Profit", data=df, hue="Customer_Gender")
```

```python
# Count plot showing the number of entries per Country, segmented by Age Group
sns.countplot(x="Country", data=df, hue="Age_Group")
```

```python
# Scatter plot of Cost vs Age Group, faceted by Product Category (columns) and Country (rows)
sns.relplot(x="Cost", y="Age_Group", data=df, kind="scatter", col="Product_Category", row="Country")
```

```python

# Line plot of Unit Cost across different States (categorical y-axis)
sns.relplot(x="Unit_Cost", y="State", data=df, kind="line")
```

```python
# Line plot of Profit over Year, with lines styled and colored by Month, using markers and solid lines
sns.relplot(x="Year", y="Profit", 
            data=df, kind="line", 
            style="Month", hue="Month", 
            markers=True, dashes=False)
```

```python
# Box plot of Revenue for each Product Category, hiding outliers for a cleaner look
sns.catplot(x="Product_Category", y="Revenue", data=df, kind="box", showfliers=False)
```

```python
# Set plot style to 'whitegrid' for better readability
sns.set_style("whitegrid")

# Point plot of average Cost for each Product Category
sns.catplot(x="Product_Category", y="Cost", data=df, kind="point")

# Rotate x-axis labels by 90 degrees to prevent overlap
plt.xticks(rotation=90)
```

```python
# Compute and visualize the correlation matrix between Cost, Profit, and Revenue
correlation = df[["Cost", "Profit", "Revenue"]]  # Select relevant numeric columns
data = correlation.corr()                        # Compute pairwise correlation
sns.heatmap(data, annot=True)                    # Plot heatmap with correlation values annotated
plt.show()                                       # Display the plot
```

```python
import plotly.express as px

# Aggregate revenue by country
country_revenue = df.groupby('Country', as_index=False)['Revenue'].sum()

# Create a choropleth map using country names and total revenue
fig = px.choropleth(country_revenue, 
                    locations='Country',                # Column with country names
                    locationmode='country names',       # Use full country names for matching
                    color='Revenue',                    # Color intensity based on total revenue
                    title='Total Revenue by Country')   # Plot title

# Display the interactive map
fig.show()
```

```python
# Show the count of records per state using a count plot with a red-blue reversed color palette
sns.catplot(x="State", 
            data=df, 
            kind="count", 
            palette="RdBu_r",   # Red to blue reversed color palette
            height=5,           # Height of the entire figure
            aspect=2)           # Width-to-height ratio (makes it wider)

# Rotate x-axis labels for better readability
plt.xticks(rotation=90)

# Display the plot
plt.show()
```

```python
# Pie chart showing the percentage distribution of Age Groups in the dataset
df["Age_Group"].value_counts().plot(kind='pie', 
                                    figsize=(14, 8), 
                                    autopct='%1.1f%%',    # Show percentages with 1 decimal place
                                    fontsize=13)          # Font size for labels

plt.title("Age Group", fontsize=14)                     # Add a title with larger font size
plt.ylabel('')                                          # Remove the default y-label for cleaner look
plt.show()
```

```python
# Plot the density distribution of the 'Profit' column
df['Profit'].plot(kind='density', figsize=(14, 6))
plt.title("Profit Density Distribution", fontsize=15)  # Corrected the title to match the data

plt.legend(["Profit Density"])  # Add a legend for clarity
plt.show()
```
